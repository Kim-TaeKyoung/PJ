```python
# 데이터 기본 정보
## data type
## missing value, missing percent
## nunique
## mean, max, min
def summary(df):
    dfs = pd.DataFrame(data = df.dtypes, columns = ['data_type'])
    dfs['missing'] = df.isnull().sum()
    dfs['nunique'] = df.nunique()
    dfs.sort_values(by = ['nunique'], inplace = True)
    return dfs
```

```python
# 1. 총 이자 비용
df['total_interest_cost'] = df['int_rate'] * df['funded_amnt']

# 2. 총 신용 한도
df['total_credit_limit'] = df['revol_bal'] / df['revol_util']

# 3. 신용카드 계좌당 평균 리볼빙 잔액
df['avg_revol_bal_per_acc'] = df['revol_bal'] / df['total_acc']

# 4. 최근 6개월간 신용 조회 수 대비 체납 횟수 비율 (체납 횟수 0을 방지하기 위해 +1을 더함)
df['inq_to_delinq_ratio'] = df['inq_last_6mths'] / (df['delinq_2yrs'] + 1)  
df.fillna(0, axis = 1, inplace = True)

df['total_funded_amnt'] = df['funded_amnt'] + df['funded_amnt_inv']
df['total_out_prncp'] = df['out_prncp'] + df['out_prncp_inv']
column = ['funded_amnt', 'funded_amnt_inv', 'out_prncp', 'out_prncp_inv']
df.drop(column, axis = 1 , inplace = True)
```

#### FE2

```python
df2 = df.copy()
```

```python
def group_rate(df):
    df['group_rate'] = 5
    df.loc[(df['int_rate'] < 0.1), 'group_rate'] = 1
    df.loc[(0.1 <= df['int_rate']) & (df['int_rate'] < 0.15), 'group_rate'] = 2
    df.loc[(0.15 <= df['int_rate']) & (df['int_rate'] < 0.2), 'group_rate'] = 3
    df.loc[(0.2 <= df['int_rate']) & (df['int_rate'] < 0.25), 'group_rate'] = 4
    df.drop('int_rate', axis = 1, inplace = True)

group_rate(df2)
df2.head(2)
```

```python
def group_dti(df):
    df['group_dti'] = 5
    df.loc[(df['dti'] < 10), 'group_dti'] = 1
    df.loc[(10 <= df['dti']) & (df['dti'] < 20), 'group_dti'] = 2
    df.loc[(20 <= df['dti']) & (df['dti'] < 30), 'group_dti'] = 3
    df.loc[(30 <= df['dti']) & (df['dti'] < 40), 'group_dti'] = 4
    df.drop('dti', axis = 1, inplace = True)

group_dti(df2)
df2.head(2)
```

```python
def group_fee(df):
    df['group_fee'] = 1
    df.loc[(0 < df['total_rec_late_fee']) & (df['total_rec_late_fee'] <= 50), 'group_fee'] = 2
    df.loc[(50 < df['total_rec_late_fee']) & (df['total_rec_late_fee'] <= 250), 'group_fee'] = 3
    df.loc[(250 < df['total_rec_late_fee']) & (df['total_rec_late_fee'] <= 500), 'group_fee'] = 4
    df.loc[(500 < df['total_rec_late_fee']) & (df['total_rec_late_fee'] <= 750), 'group_fee'] = 5
    df.loc[(750 < df['total_rec_late_fee']), 'group_fee'] = 6
    df.drop('total_rec_late_fee', axis = 1, inplace = True)

group_fee(df2)
df2.head(2)
```

```python
def test_data():
    test = pd.read_csv('data/test.csv')
    test.drop(['ID', 'purpose4'], axis = 1, inplace = True)
    df['total_interest_cost'] = df['int_rate'] * df['funded_amnt']
    df['total_credit_limit'] = df['revol_bal'] / df['revol_util']
    df['avg_revol_bal_per_acc'] = df['revol_bal'] / df['total_acc']
    df['inq_to_delinq_ratio'] = df['inq_last_6mths'] / (df['delinq_2yrs'] + 1)  
    test.fillna(0, axis = 1, inplace = True)

    test['total_funded_amnt'] = test['funded_amnt'] + test['funded_amnt_inv']
    test['total_out_prncp'] = test['out_prncp'] + test['out_prncp_inv']
    column = ['funded_amnt', 'funded_amnt_inv', 'out_prncp', 'out_prncp_inv']
    test.drop(column, axis = 1 , inplace = True)
    group_rate(test)
    group_dti(test)
    group_fee(test)
    return test
```

### stacking

```python
from sklearn.model_selection import train_test_split, KFold
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from lightgbm import  LGBMClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from autogluon.tabular import TabularDataset, TabularPredictor
import h2o
from h2o.automl import H2OAutoML
from sklearn.metrics import accuracy_score
```

```python
estimators = []
estimators.append(('Decision Tree Classifier', DecisionTreeClassifier(random_state=42) ))
estimators.append(('Gradient Boosting Classifier', GradientBoostingClassifier(random_state=42) ))
estimators.append(('Hist Gradient Boosting Classifier', HistGradientBoostingClassifier(random_state=42) ))
estimators.append(('KNN', KNeighborsClassifier() ))
estimators.append(('LogisticRegression', LogisticRegression(max_iter=1000, random_state=42)))
estimators.append(('RandomForest', RandomForestClassifier(max_depth= 10, min_samples_leaf= 1, min_samples_split= 3, n_estimators= 170, random_state=42) ))
estimators.append(('XGB', XGBClassifier(**xgb_params) ))
estimators.append(('CatBoost', CatBoostClassifier(**cat_params) ))
estimators.append(('lgbm', LGBMClassifier(**lgbm_params) ))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score

skf = StratifiedKFold(n_splits=5)
scaler = StandardScaler()
f1_scores = []
test_preds = np.empty((5, len(test)))

for i, (train_idx, valid_idx) in enumerate(skf.split(X, y)):
    
    X_train, X_valid = X.iloc[train_idx], X.iloc[valid_idx]
    y_train, y_valid = y.iloc[train_idx], y.iloc[valid_idx]
    
    scaler.fit(X_train)

    X_train = scaler.transform(X_train)
    X_valid = scaler.transform(X_valid)
    
    test = test_data()
    test = scaler.transform(test)
    
    stacking_clf = StackingClassifier(estimators=estimators, final_estimator=xgb)
    stacking_clf.fit(X_train, y_train)
    y_pred = stacking_clf.predict(X_valid)
    
    f1 = f1_score(y_valid, y_pred, average='macro')
    f1_scores.append(f1)
    
    test_pred = stacking_clf.predict(test)
    test_preds[i, : ] = test_pred
    
print(np.mean(f1_scores))  
```
